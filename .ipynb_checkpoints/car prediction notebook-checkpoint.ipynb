```python
# since there is no use of car name in predicting the price so we would remove it from here
df.columns
```

```python
# in this step we will remove the car name from the columns
final_dataset=df[['year', 'selling_price', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner']]
```

```python
final_dataset.head()
```

```python
# now we will add new feature i.e. current year so that after subtracting current year and year column we can get the nunmber of years for which car has been used
final_dataset["current_year"]=2020
```

```python
final_dataset.head()
```

```python
final_dataset["no_years"]=final_dataset["current_year"]-final_dataset["year"]
```

```python
final_dataset.head()
```

```python
final_dataset.drop(["year"],axis=1,inplace=True) # axis=1 for columns and iplace=true to change values in actual dataframe ie final_dataset here
```

```python
final_dataset.head()
```

```python
final_dataset.drop(["current_year"],axis=1,inplace=True)
```

```python
final_dataset
```

```python
final_dataset=pd.get_dummies(final_dataset,drop_first=True) # here we have set dummy variable for the categorical variables . also we have set categorical variable to true so that suppose for fuel first varible value will be dropped i.e. cng here and other four will be set values
```

```python
final_dataset.head() # here in data as we can see that fuel_cng is not in data so if any of the fuels does not have value=1 than it will be cng fuel 
# it is used to avoid dummy variable trap
```

```python
# now we will find correlation
final_dataset.rename(columns={'owner_Fourth & Above Owner': "Fourth_and_above", "owner_Second Owner": 'second_owner',"owner_Test Drive Car":"test_drive_car","owner_Third Owner":"third_owner","seller_type_Trustmark Dealer":"seller_Trustmark_dealer"},inplace=True)
final_dataset.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
#sns.pairplot(final_dataset)
```

```python
# feature selection
corrmat=final_dataset.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(20,20))
g=sns.heatmap(final_dataset[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# x is independent feature and y is dependent feature
x=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]
final_dataset.head()
```

```python
x.head()
```

```python
y.head()
```

```python
# Now we will check for  feature importance ie the features which are the most important
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_) # feature importances gives the values of features which are the most important
```

```python
feat_importances=pd.Series(model.feature_importances_,index=x.columns) # pd.series will convert model into series
feat_importances.nlargest(5).plot(kind="barh") # it will plot the 5 most important features on a bargraph ,h here stands for horizontal graph
plt.show() # but here we will consider all the features in our model
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
### Hyperparameters
import numpy as np
# n_estimators is used for making decision trees of some length .It gives us number of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=100, stop=1200,num=12)] # here values of n_estimators lies between 100 and 1200 its default value is 100
print(n_estimators)
```

```python
# number of features to consider at every split
max_features=["auto","sqrt"]
# maximum number of levels in the tree
max_depth=[int(x) for x in np.linspace(start=5,stop=30,num=6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split=[2,5,10,15,100]
# minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV # randomizedsearchcv helps us to find the values of parameters best suited
```

```python
random_grid={'n_estimators':n_estimators,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
rf=RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring="neg_mean_squared_error",n_iter=10,cv=5,verbose=1,random_state=42,n_jobs=1)
# here estimator is randomforestregressor,no of t=iterations=10,crossvalidation=5,verbose=1 means all the cores of laptop will be used
```

```python
rf_random.fit(x_train,y_train)
```

```python
predictions=rf_random.predict(x_test)
predictions
```

```python
# now we have got the values  of predictions our job is to compare the values of predictions with y_test
# for this purpose we will use distplot
sns.distplot(y_test-predictions) # more narrow graph at the middle here means that our model is predicting near to accurate so model is good to go
```

```python
plt.scatter(y_test,predictions)
```

```python
# creating the pickle file
#so from  here from the graph we can observe that the graph is pretty much linear so our predictions are good
import pickle
file=open("random_forest_regression_model_original1.pkl","wb") # here we are opening the file in writebyte modeor
# dump information to that file
pickle.dump(rf_random,file)
```
